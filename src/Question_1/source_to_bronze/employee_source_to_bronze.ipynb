```python
employee_path = "/Volumes/workspace/default/employee_raw_vol/Employee-Q1.csv"

department_path = "/Volumes/workspace/default/employee_raw_vol/Department-Q1.csv"

country_path = "/Volumes/workspace/default/employee_raw_vol/Country-Q1.csv"
```

```python
employee_df = spark.read.format("csv") \
.option("header","true") \
.option("inferSchema","true") \
.load(employee_path)

department_df = spark.read.format("csv") \
.option("header","true") \
.option("inferSchema","true") \
.load(department_path)

country_df = spark.read.format("csv") \
.option("header","true") \
.option("inferSchema","true") \
.load(country_path)
```

```python
display(employee_df)
display(department_df)
display(country_df)
```

```python
employee_df.write.mode("overwrite") \
.option("header","true") \
.csv("/Volumes/workspace/default/employee_raw_vol/source_to_bronze/employee")

department_df.write.mode("overwrite") \
.option("header","true") \
.csv("/Volumes/workspace/default/employee_raw_vol/source_to_bronze/department")

country_df.write.mode("overwrite") \
.option("header","true") \
.csv("/Volumes/workspace/default/employee_raw_vol/source_to_bronze/country")
```

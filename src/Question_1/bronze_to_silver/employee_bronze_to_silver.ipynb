```python
employee_bronze_path = "/Volumes/workspace/default/employee_raw_vol/source_to_bronze/employee"

department_bronze_path = "/Volumes/workspace/default/employee_raw_vol/source_to_bronze/department"

country_bronze_path = "/Volumes/workspace/default/employee_raw_vol/source_to_bronze/country"
```

```python
from pyspark.sql.types import *

employee_schema = StructType([
    StructField("EmployeeID", IntegerType(), True),
    StructField("EmployeeName", StringType(), True),
    StructField("DepartmentID", IntegerType(), True),
    StructField("CountryID", IntegerType(), True),
    StructField("Age", IntegerType(), True),
    StructField("Salary", DoubleType(), True)
])
```

```python
employee_df = spark.read.format("csv") \
.option("header","true") \
.schema(employee_schema) \
.load(employee_bronze_path)
```

```python
display(employee_df)
```

```python
employee_df = convert_to_snake_case(employee_df)
```

```python
employee_df.columns
```

```python
employee_df = add_load_date(employee_df)
```

```python
spark.sql("CREATE DATABASE IF NOT EXISTS Employee_info")
```

```python
silver_path = "dbfs:/Volumes/workspace/default/employee_raw_vol/silver/Employee_info/dim_employee"
```
